# **Exploring Crime in LA**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
LAData= pd.read_csv("C:/Users/Regina/Documents/GitHub/DSCI343Project/awm48/SortedData/LosAngelesAssault.csv")
```

```python
dataframe = LAData.sample(frac=1)
```

```python
dataframe['Datetime Occurred'] =  pd.to_datetime(dataframe['Date Occurred'], format='%m/%d/%Y')
print(len(dataframe['Datetime Occurred']))
```

```python
plt.title("Uniform Distribution of # of assault crimes per year")
plt.hist(dataframe['Datetime Occurred'], 8)
plt.show()
```

```python
time = pd.DataFrame()
time['weekdays'] = dataframe['Datetime Occurred'].apply(lambda x: x.weekday())
time['year'] = dataframe['Datetime Occurred'].dt.year
time['month'] = dataframe['Datetime Occurred'].dt.month
time['day'] = dataframe['Datetime Occurred'].dt.day
time['timeDay'] = dataframe['Time Occurred']
time['Area ID'] = dataframe['Area ID']

plt.title("Distribution of # of assault crimes by day of week")
plt.hist(time['weekdays'], 7)
plt.show()
```

```python
plt.title("Distribution of of assault crimes per month")
plt.hist(time['month'], 12)
plt.show()
```

```python
plt.title("Distribution of of assault crimes by day of month")
plt.hist(time['day'], 31)
plt.show()
```

```python
plt.title("Distribution of # of assault crimes by hour")
plt.hist(dataframe['Time Occurred'], 24)
plt.show()
```
