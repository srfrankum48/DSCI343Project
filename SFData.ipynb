# **Exploring Crime in San Francisco**

```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SFData = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv", error_bad_lines=False)
```

```python
dataframe = SFData.sample(frac=1)
```

```python
dataframe['Datetime Occurred'] =  pd.to_datetime(dataframe['Date'] + ' ' + dataframe['Time'], format='%m/%d/%Y %H:%M')
print(len(dataframe['Datetime Occurred']))
```

```python
plt.title("Uniform Distribution of # crimes per year")
plt.hist(dataframe['Datetime Occurred'], 8)
plt.show()
```

```python
time = pd.DataFrame()
time['weekdays'] = dataframe['DayOfWeek']
time['year'] = dataframe['Datetime Occurred'].dt.year
time['month'] = dataframe['Datetime Occurred'].dt.month
time['day'] = dataframe['Datetime Occurred'].dt.day
time['timeDay'] = dataframe['Datetime Occurred'].dt.hour
time['Area ID'] = dataframe['PdDistrict']

plt.title("Distribution of # crimes by day of week")
plt.hist(time['weekdays'], 7)
plt.show()
```

```python
plt.title("Distribution of crimes per month")
plt.hist(time['month'], 12)
plt.show()
```

```python
plt.title("Distribution of crimes by day of month")
plt.hist(time['day'], 31)
plt.show()
```

```python
plt.title("Distribution of # crimes by hour")
plt.hist(time['timeDay'], 24)
plt.show()
```
